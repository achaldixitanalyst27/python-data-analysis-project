```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import re

# Function to clean text
def clean_text(text):
    text = text.lower()
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
    tokens = word_tokenize(text, preserve_line=True)  # Fixed: preserve_line=True avoids sent_tokenize error
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return " ".join(tokens)
```

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer

# Dummy data with user_review column
df = pd.DataFrame({
    'title': ['Movie A', 'Movie B', 'Movie C'],
    'user_review': [
        'Absolutely loved it! A masterpiece.',
        'Not bad, but not great either.',
        'Worst movie ever. Complete waste of time.'
    ]
})

# Sentiment analysis
sid = SentimentIntensityAnalyzer()
df['sentiment_score'] = df['user_review'].apply(lambda x: sid.polarity_scores(str(x))['compound'])
df['sentiment_label'] = df['sentiment_score'].apply(
    lambda score: 'positive' if score > 0.05 else ('negative' if score < -0.05 else 'neutral')
)

print(df[['title', 'user_review', 'sentiment_score', 'sentiment_label']])
```

```python
# Count sentiment labels
import matplotlib.pyplot as plt

sentiment_summary = df['sentiment_label'].value_counts()

print(sentiment_summary)

# Plot pie chart
sentiment_summary.plot(
    kind='pie',
    colors=['green', 'red', 'gray'],  # make sure this order matches sentiment_summary.index
    autopct='%1.1f%%',  # show percentage on slices
    startangle=90,
    legend=False  # disable automatic legend so we can add custom one
)

plt.title('Sentiment Distribution in User Reviews')
plt.ylabel('')  # hide ylabel for clean look

# Add legend with labels and colors matching the pie slices
plt.legend(
    labels=sentiment_summary.index,
    loc='best',
    title='Sentiment',
    fontsize='small'
)

plt.show()
```

```python
movie_sentiment = df.groupby('title')['sentiment_score'].mean().reset_index()
top_positive = movie_sentiment.sort_values(by='sentiment_score', ascending=False).head(10)
print(top_positive)
```

```python
# Step 4


# Plot genres
plt.figure(figsize=(10,8))
genre_counts.head(15).plot(kind='bar', color='Yellow')
plt.title('Top 15 Genres by Number of Titles')
plt.ylabel('Number of Titles')
plt.xlabel('Genre')
plt.xticks(rotation=90)
plt.show()

# Plot types
plt.figure(figsize=(6,4))
type_counts.plot(kind='bar', color='orange')
plt.title('Titles by Content Type')
plt.ylabel('Number of Titles')
plt.xlabel('Type')
plt.xticks(rotation=20)
plt.show()
```

```python


# Pie chart for top 15 genres
plt.figure(figsize=(7,7))
genre_counts.head(15).plot(kind='pie', autopct='%1.1f%%', startangle=140, shadow=True)
plt.title('Top 15 Genres by Number of Titles')
plt.ylabel('')  # Hide y-label for pie chart
plt.axis('equal')  # Equal aspect ratio to make the pie circular
plt.show()

# Pie chart for types (Movie, TV Show)
plt.figure(figsize=(6,7))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['orange', 'lightgreen'], shadow=True)
plt.title('Titles by Content Type')
plt.ylabel('')
plt.axis('equal')
plt.show()
```
