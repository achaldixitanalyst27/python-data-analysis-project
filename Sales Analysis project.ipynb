Data Visualization 

```python
# creating list of numeric columns with year excluded
numeric_cols = df.select_dtypes(include=['float64']).columns.to_list()[1:]
numeric_cols
```

```python
# distribution of sales data 
fig, axes = plt.subplots(2,3,figsize=(15,10))
colors =[]
n=0
for i in range(2):
    for j in range(3):
        if n>4:
            break
        sns.boxenplot(df[numeric_cols[n]], fill=True, ax=axes[i,j])
        axes[i,j].set_title(f"{numeric_cols[n]}")
        n+=1
plt.show()
```

```python
sns.heatmap(df[numeric_cols].corr(numeric_only=True), annot=True)
plt.title("Sales Correlation Heatmap",fontdict={'size':16})
plt.xticks(rotation=80)
plt.show()
```

What are the Top 10 Game publication globally ?

```python
top10_publisher = df.groupby('Publisher')['Global_Sales (in millions)'].sum().sort_values(ascending=False)[:10]
top10_publisher
```

```python
plt.pie(top10_publisher.values, labels=top10_publisher.index, autopct='%.2f', colors=plt.get_cmap('Set2').colors)
plt.title('Biggest publishers int the vedio gamesindustry', fontdict= {'weight': 'bold'})
plt.legend(bbox_to_anchor=(1.4, 1.1))
plt.show()
```

```python
for i in numeric_cols:
    top10_publisher = df.groupby('Publisher')[i].sum().sort_values(ascending=False).head(10)
    plt.pie(top10_publisher.values, labels=top10_publisher.index, autopct='%.2f', colors=plt.get_cmap('Paired').colors)
    plt.title(f"Biggest publishers in the vedio gamesindustry {i}", fontdict= {'weight': 'bold'})
    plt.legend(bbox_to_anchor=(1.4, 1.1))
    plt.show()
```

Which Publication has released the most number of games?

```python
top10_numofgames = df['Publisher'].value_counts().sort_values(ascending=False)[:10]
top10_numofgames
```

```python
colors = plt.cm.brg(np.linspace(0,1, len(top10_numofgames.index)))
bars = plt.bar(top10_numofgames.index, top10_numofgames.values, color=colors)
plt.xticks(rotation=90)
ax = plt.gca()  # Get the current axes
xticks = ax.get_xticklabels()
for tick, color in zip(xticks, colors):
    tick.set_color(color)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2 , height,f'{height}',
             ha='center', va='bottom', fontsize=10 , color='navy')
plt.title("Top 10 Publications by No. of games released", fontdict={'weight':'bold'}, color='tab:blue')
plt.show()
```

Which video Game had the most sales?

```python
top_10_vg = df.sort_values('Global_Sales (in millions)',ascending=False)[:10]
top_10_vg
```

```python
plt.barh(top_10_vg['Name'].values[::-1],top_10_vg["Global_Sales (in millions)"].values[::-1],
                   color=plt.get_cmap('Dark2').colors)
plt.title("top 10 vedio games of past 4 decades")
plt.show()
```

Most Sucessful Game of Each Year(Creation Year)

```python
best_each_year = df.loc[df.groupby('Year')['Global_Sales (in millions)'].idxmax()]
best_each_year
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(best_each_year['Year'].values)))
plt.figure(figsize=(14,16))
bars =plt.bar(best_each_year['Year'].values,best_each_year['Global_Sales (in millions)'].values, color=colors)
n =0
for bar in bars:
    x = bar.get_x()
    y=bar.get_height()
    plt.text(x+0.5 , y+0.5, best_each_year['Name'].values[n],ha='center',va='bottom',rotation=90)
    n+=1
plt.title("Most Succesful Game (from Each Year)",pad=20)
plt.show()
```

Video Game by Genre

```python
games_genre = df['Genre'].value_counts()
games_genre
```

```python
colors = plt.cm.cool(np.linspace(0,1, len(games_genre.index)))
explode = tuple(0.01 for i in range (12))
plt.pie(games_genre.values , labels=games_genre.index, autopct='%2f',colors=colors,pctdistance=0.85,
        explode=explode, wedgeprops={'width':0.3})
plt.title('Video Games By Gerne')
plt.show()
```
