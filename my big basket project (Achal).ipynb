```python
# Fill missing values in 'rating' column with the average rating
# Replace missing values in 'description' column with an empty string

mean_rating = df['rating'].mean()
df['rating'] = df['rating'].fillna(mean_rating)
df['description'] = df['description'].fillna('')
```

```python

# we have sort the table how much raing of the product those have best rating they should bemost selling products

# Top 5 best selling products
top_sold = df.sort_values(by='rating', ascending=False).head(5)
print("Top 5 Sold Products:")
print(top_sold[['product', 'rating']])

# Least 5 sold products
least_sold = df.sort_values(by='rating', ascending=True).head(5)
print("\nLeast 5 Sold Products:")
print(least_sold[['product', 'rating']])
```

```python
# Replace missing values in 'rating' with the column's mean value to avoid gaps
# Replace missing values in 'description' with an empty string to handle missing text

print(df[['product', 'market_price', 'sale_price', 'discount']].head())
df['discount'] = df['market_price'] - df['sale_price']
```

```python
# check the outliers anf if there then convert with mean

numeric_cols = ['sale_price', 'market_price', 'rating']

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in '{col}':", len(outliers))

    mean_val = df[col].mean()
    df.loc[(df[col] < lower_bound) | (df[col] > upper_bound), col] = mean_val
```

```python
# Plot Histogram (Distribution of a Numeric Column)
plt.figure(figsize=(8,5))
sns.histplot(df['rating'], bins=30, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot to visualize the spread of 'sale_price' and identify outliers in the data.

plt.figure(figsize=(8,5))
sns.boxplot(x=df['sale_price'])
plt.title('Boxplot of Sale Price')
plt.show()
```

```python
#  Countplot to show the frequency of products in each 'category'.  
# Rotating x-axis labels to make category names readable.

plt.figure(figsize=(10,6))
sns.countplot(x='category', data=df)
plt.title('Count of Products by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#  Scatter plot to observe the relationship between 'market_price' and 'sale_price'.

plt.figure(figsize=(8,6))
sns.scatterplot(x='market_price', y='sale_price', data=df)
plt.title('Market Price vs Sale Price')
plt.show()
```

```python
# Heatmap to display correlation coefficients between numeric columns, helping identify which variables are strongly related.

plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Bar plot to show average 'sale_price' for each product 'category'.  
# X-axis labels are rotated for better readability.

plt.figure(figsize=(10,6))
sns.barplot(x='category', y='sale_price', data=df)
plt.title('Average Sale Price by Category')
plt.xticks(rotation=45)
plt.show()
```
